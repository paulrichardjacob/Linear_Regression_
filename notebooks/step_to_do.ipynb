### Missing values Existing why?
- MCAR (missing completely at random) — e.g., a sensor glitch. Safe to drop or impute simply.
- MAR (missing at random, but related to other observed columns) — e.g., income missing more often for younger respondents. Imputation should condition on those related columns.
- MNAR (missing not at random) — e.g., people with high income refuse to report it. This is the dangerous one — the missingness itself carries information, and naive imputation will bias your results.
### Handling the Missing Values
- Overall % is under 5% simple imputation(Mean/Median/Mode) or Row-drop is at low risk
- 5 -30% missing Check if the missingness corelates with any other columns by using groupby like this specific group is missing the data point. if yes try to impute based on specific group or create a new category 'Unknown'
- 30-60% missing - It depends on whethere the column is important to the target. if it is a weak predictor, drop the column. if it is important consider add it as an another category like 'Unknown'. which preseves the feature importance.
- Over 60% missing in a column - Drop the column outright unless you have strong domain reason to believe the missingness itself is the signal, in which case keep it as a categorical "missing/present" indicator and drop the raw values.

# What to do, once outliers found

- Step 1:
    - First classify which category it belongs
        - Data Entry error - ex: age= 999, negative price, These are a bug not a real value
        - Legitimate extreme value - ex: billionare income in a income dataset, a mansion price in housing price dataset. They are Real but just rare
        - Measurement/process artifact - ex: sensor spike , a one time glitch
- Step 2:
    - Data Entry error: 
        - Fix if you can figure out the true value, other wise treat it as missing value and treat with missing work-flow
    - Legitimate extreme values: 
        - Usually keep them but consider whether your model is sensitive to outlier like Regression model , KNN, K-means. if they are Tree based model leave it alone as it it robust.
    - Measurement artifact:
        - Remove that row or cap it

# Situation if the outliers are ligit and want to use outlier sensitive model 
- Cap it - clip the values beyond a percentile (e.g., 1st/99th) instead of deleting the row. Keeps the row, tames the extremity.
- Transform the column - log transfor or many more is the standard move for right skewed data with legitimate high-end outliers. 
- Drop the row - only justified if you've confirmed it's an error, not just because it's convenient.
